```

import pandas as pd
import nltk

nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

import csv
csv.field_size_limit(2147483647)

df = pd.read_csv('train1.csv', engine='python')

df.iloc[0]

text = df.iloc[0]['text']
text = text.lower()

from nltk import word_tokenize, sent_tokenize

sentences = sent_tokenize(text)
print(sentences)

words = word_tokenize(text)
print(words)

from nltk.corpus import stopwords
stop_words = stopwords.words('english')

filtered = []
for word in words:
    if word.lower() not in stop_words:
        filtered.append(word)

print(filtered)

cleaned = [word for word in filtered if word.isalnum()]
print(cleaned)

nltk.download('averaged_perceptron_tagger_eng')
from nltk import pos_tag

pos_tags = pos_tag(cleaned)
print(pos_tags)

from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer('english')
stemmed = [stemmer.stem(word) for word in cleaned]
print(stemmed)

from nltk.stem import WordNetLemmatizer

lm = WordNetLemmatizer()
lemmatized = [lm.lemmatize(word) for word in cleaned]
print(lemmatized)

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([' '.join(lemmatized)])
print(f"Shape of TF-IDF matrix: {tfidf_matrix.shape}")

corpus = df['text']

vectorizer = TfidfVectorizer(max_features=5000)

from collections import Counter

tf = Counter(cleaned)
total_words = len(cleaned)

tf_normalized = {word: count/total_words for word, count in tf.items()}
print("Term Frequency:\n", tf_normalized)

import math

documents = corpus.astype(str).tolist()
N = len(documents)

word_doc_count = {}

for doc in documents:
    words = set(doc.split())
    for word in words:
        word_doc_count[word] = word_doc_count.get(word, 0) + 1

idf = {word: math.log(N / count) for word, count in word_doc_count.items()}
print("IDF:\n", dict(list(idf.items())[:20]))

tfidf_matrix = vectorizer.fit_transform(documents)
print(tfidf_matrix.shape)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)

print("TF-IDF Matrix:\n")
print(tfidf_df)
```
